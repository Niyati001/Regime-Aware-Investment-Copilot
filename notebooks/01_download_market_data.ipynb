```python
import yfinance as yf
from pathlib import Path

RAW_DIR = Path("../data/raw")
RAW_DIR.mkdir(parents=True, exist_ok=True)

tickers = {
    "^NSEI": "NIFTY_50",
    "^GSPC": "SP500",
    "^IXIC": "NASDAQ",
    "^VIX": "VIX"
}

data = yf.download(
    list(tickers.keys()),
    start="2010-01-01",
    end="2026-01-01",
    auto_adjust=True,
    group_by="column"
)

data.to_csv(RAW_DIR / "market_data.csv")

print("Saved market data to:", RAW_DIR / "market_data.csv")
print(data.shape)
data.head()
```
