This notebook summarizes the full project into a research-style report.

Project question:

> Can market regime detection improve financial prediction compared with a single model trained across all market conditions?

## Executive Summary

This project builds a regime-aware market prediction system.

The system has three main stages:

1. **Feature engineering** from market prices and volatility data.
2. **Hidden Markov Model (HMM)** to detect latent market regimes.
3. **Mixture of Experts (MoE)** to train specialized models for different regimes.

The main finding is that the regime-aware MoE classifier improved walk-forward performance over a baseline classifier.

## 4. HMM Regime Detection

The Hidden Markov Model tries to discover hidden states from observable market features.

Observed features:

- Return
- Volatility
- Momentum
- Drawdown
- VIX

Hidden states:

- Bull
- Bear
- Crisis

The HMM does not know these labels in advance. It learns numerical states, then we interpret them using return, volatility, and drawdown behavior.

````python
regime_summary = (
    regimes
    .groupby("regime_label")
    .agg(
        observations=("regime_label", "size"),
        avg_daily_return=("return_1d", "mean"),
        annualized_return=("return_1d", lambda x: x.mean() * 252),
        annualized_volatility=("return_1d", lambda x: x.std() * np.sqrt(252)),
        avg_momentum_30d=("momentum_30d", "mean"),
        avg_drawdown=("drawdown", "mean"),
        worst_drawdown=("drawdown", "min"),
        avg_vix=("vix_close", "mean"),
    )
    .sort_values("annualized_return", ascending=False)
)

regime_summary
````

````python
colors = {
    "Bull": "green",
    "Bear": "orange",
    "Crisis": "red",
}

plt.figure(figsize=(14, 6))
plt.plot(regimes["Date"], regimes["Close"], color="black", linewidth=1, label="Close")

for label, color in colors.items():
    temp = regimes[regimes["regime_label"] == label]
    plt.scatter(temp["Date"], temp["Close"], s=10, color=color, label=label, alpha=0.7)

plt.title("HMM-Detected Market Regimes")
plt.xlabel("Date")
plt.ylabel("Close")
plt.legend()
plt.show()
````

## 8. Out-of-Sample Equity Curves

The strategy is deliberately simple:

```text
If predicted probability of positive return >= 0.45, hold the index.
Otherwise, stay in cash.
```

This is not a final trading strategy, but it is useful for checking whether predictions have practical directional value.

````python
def strategy_returns(actual_returns, up_prob, threshold=0.45):
    signal = (np.asarray(up_prob) >= threshold).astype(int)
    return signal * np.asarray(actual_returns)

threshold = 0.45

walk_predictions = walk_predictions.sort_values("Date").reset_index(drop=True)

equity_df = walk_predictions[["Date", "target_return_1d"]].copy()
equity_df["buy_hold"] = (1 + equity_df["target_return_1d"]).cumprod()
equity_df["baseline"] = (1 + strategy_returns(walk_predictions["target_return_1d"], walk_predictions["baseline_up_prob"], threshold)).cumprod()
equity_df["hard_moe"] = (1 + strategy_returns(walk_predictions["target_return_1d"], walk_predictions["hard_moe_up_prob"], threshold)).cumprod()
equity_df["soft_moe"] = (1 + strategy_returns(walk_predictions["target_return_1d"], walk_predictions["soft_moe_up_prob"], threshold)).cumprod()

plt.figure(figsize=(14, 6))
plt.plot(equity_df["Date"], equity_df["buy_hold"], label="Buy & Hold", color="black")
plt.plot(equity_df["Date"], equity_df["baseline"], label="Baseline")
plt.plot(equity_df["Date"], equity_df["hard_moe"], label="Hard MoE")
plt.plot(equity_df["Date"], equity_df["soft_moe"], label="Soft MoE")
plt.title("Walk-Forward Out-of-Sample Equity Curves")
plt.xlabel("Date")
plt.ylabel("Growth of $1")
plt.legend()
plt.show()
````

## 9. Final Findings

Based on the walk-forward results:

- The regime-aware MoE achieved higher average accuracy than the baseline.
- The regime-aware MoE achieved higher average F1 score than the baseline.
- The regime-aware MoE achieved higher average ROC-AUC than the baseline.
- The regime-aware MoE achieved higher average strategy Sharpe than the baseline.

This supports the core research hypothesis:

> Market regimes contain useful information, and regime-aware expert models can improve directional prediction compared with a single model trained across all market states.

## 10. Limitations

This is a strong student/research prototype, but it is not production trading infrastructure.

Limitations:

- Current full pipeline uses S&P 500 first; all-asset modeling is a future extension.
- Transaction costs are not included.
- Slippage is not included.
- HMM regime labels are inferred, not ground truth.
- The dataset is still small for deep learning.
- Random Forest is used for interpretability and stability; XGBoost/LightGBM may improve results.
